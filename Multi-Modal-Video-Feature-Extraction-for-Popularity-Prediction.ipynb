```python
import os
import torch
import pandas as pd
from torch.utils.data import Dataset, DataLoader
import torch.nn as nn
import numpy as np
import random
target_name_list=['video_share_count','video_heart_count','video_comment_count','video_play_count']
# Set seed for reproducibility
def set_seed(seed=42):
    random.seed(seed)
    np.random.seed(seed)
    torch.manual_seed(seed)
    if torch.cuda.is_available():
        torch.cuda.manual_seed(seed)
        torch.cuda.manual_seed_all(seed)
    torch.backends.cudnn.deterministic = True
    torch.backends.cudnn.benchmark = False

# Call set_seed at the start
set_seed(42)


# Load training and testing data
train_excel_file = '/content/inform2024/training.xlsx'
test_excel_file = '/content/inform2024/offical_test.xlsx'
#test_excel_file = '/content/test.xlsx'
df_train = pd.read_excel(train_excel_file)
df_test = pd.read_excel(test_excel_file)

# Filter out invalid video_ids from df_test
# invalid_video_ids = [
    # '6927569510555798786', '6954318760312360193', '6956867508397788418',
    # '6991387228106738945', '6997796531906628869', '7008440765454044418',
    # '7028958462516268293', '7038771289485561135', '7059479232673615151'
# ]

# Clean video_id and apply transformation
df_train['video_id'] = df_train['video_id'].astype(str).str.replace("'", "")
df_test['video_id'] = df_test['video_id'].astype(str).str.replace("'", "")
#df_test = df_test[~df_test['video_id'].isin(invalid_video_ids)]
df_train[target_name_list] = np.log1p(df_train[target_name_list])
df_test[target_name_list] = np.log1p(df_test[target_name_list])
# Define 5 different feature folder paths for training and testing
train_folder_paths = [
    '/content/videomae_features_training', '/content/vivit_features-training',
    '/content/timesformer-features-training', '/content/xclip-large-patch14-kinetics-600-training',
    '/content/opengv_bert_training',
    '/content/lava_bert_training'

]

test_folder_paths = [
    '/content/videomae_features_testing', '/content/vivit_features-testing',
    '/content/timesformer-features-testing', '/content/xclip-large-patch14-kinetics-600-testing',
    '/content/opengv_bert_testing',
    '/content/lava_bert_testing'

]
# Custom dataset to return 5 features and 1 target
class VideoDataset(Dataset):
    def __init__(self, dataframe, folder_paths, device='cpu',target_name='video_share_count'):
        self.dataframe = dataframe
        self.folder_paths = folder_paths  # Paths for 5 different folders
        self.device = device  # Store device info
        self.targets = self.dataframe[target_name].values.reshape(-1, 1)

    def __len__(self):
        return len(self.dataframe)

    def __getitem__(self, idx):
        video_id = self.dataframe.iloc[idx]['video_id']

        try:
            # Load features from 5 separate folders
            feature_tensors = []
            for folder in self.folder_paths:
                file_path = os.path.join(folder, f'{video_id}_repaired.pt')
                #print(file_path)
                feature_tensor = torch.load(file_path)
                feature_tensor = torch.tensor(feature_tensor).to(self.device)
                feature_tensors.append(feature_tensor)

            # Unpack the feature tensors into separate variables
            feature1, feature2, feature3, feature4, feature5, feature6 = feature_tensors

            # Get the target value
            target = self.targets[idx]
            target_tensor = torch.tensor(target, dtype=torch.float32).to(self.device)
            # print(target_tensor)

            return feature1, feature2, feature3, feature4, feature5, feature6, target_tensor

        except FileNotFoundError as e:
            #print(f"Error: Could not find pt files for video_id {video_id}. Skipping this sample.")
            return None

# Neural network model
import torch.nn.functional as F

class VideoDataset_pred(Dataset):
    def __init__(self, dataframe, folder_paths, device='cpu',target_name='video_share_count'):
        self.dataframe = dataframe
        self.folder_paths = folder_paths
        self.device = device
        self.targets = self.dataframe[target_name].values.reshape(-1, 1)

    def __len__(self):
        return len(self.dataframe)

    def __getitem__(self, idx):
        video_id = self.dataframe.iloc[idx]['video_id']

        try:
            # Load features from 5 separate folders
            feature_tensors = []
            for folder in self.folder_paths:
                file_path = os.path.join(folder, f'{video_id}_repaired.pt')
                feature_tensor = torch.load(file_path)
                feature_tensor = torch.tensor(feature_tensor).to(self.device)
                feature_tensors.append(feature_tensor)

            feature1, feature2, feature3, feature4, feature5, feature6 = feature_tensors

            target = self.targets[idx]
            target_tensor = torch.tensor(target, dtype=torch.float32).to(self.device)

            return feature1, feature2, feature3, feature4, feature5, feature6, target_tensor, video_id

        except FileNotFoundError as e:
            return None
```

```python
# Custom collate function to handle 5 features and target
def collate_fn(batch):
    batch = list(filter(lambda x: x is not None, batch))
    if not batch:
        return (torch.empty(0),) * 6

    feature1_list, feature2_list, feature3_list, feature4_list, feature5_list, feature6_list, targets_list = zip(*batch)

    feature1 = torch.stack(feature1_list)
    feature2 = torch.stack(feature2_list)
    feature3 = torch.stack(feature3_list)
    feature4 = torch.stack(feature4_list)
    feature5 = torch.stack(feature5_list)
    feature6 = torch.stack(feature6_list)
    targets = torch.stack(targets_list)

    return feature1, feature2, feature3, feature4, feature5, feature6, targets
def collate_fn_pred(batch):
    batch = list(filter(lambda x: x is not None, batch))
    if not batch:
        return (torch.empty(0),) * 7

    feature1_list, feature2_list, feature3_list, feature4_list, feature5_list, feature6_list, targets_list, video_id_list = zip(*batch)

    feature1 = torch.stack(feature1_list)
    feature2 = torch.stack(feature2_list)
    feature3 = torch.stack(feature3_list)
    feature4 = torch.stack(feature4_list)
    feature5 = torch.stack(feature5_list)
    feature6 = torch.stack(feature6_list)
    targets = torch.stack(targets_list)
    video_ids = list(video_id_list)

    return feature1, feature2, feature3, feature4, feature5, feature6, targets, video_ids
# Prepare dataloaders to return 5 features and 1 target
def prepare_dataloaders(df, folder_paths, batch_size=32, device='cpu',target_name='video_share_count'):
    dataset = VideoDataset(df, folder_paths, device=device,target_name=target_name)
    data_loader = DataLoader(dataset, batch_size=batch_size, shuffle=True, collate_fn=collate_fn)
    return data_loader

def prepare_dataloaders_pred(df, folder_paths, batch_size=32, device='cpu',target_name='video_share_count'):
    dataset = VideoDataset_pred(df, folder_paths, device=device,target_name=target_name)
    data_loader = DataLoader(dataset, batch_size=batch_size, shuffle=False, collate_fn=collate_fn_pred)
    return data_loader
# MAPE calculation function
def calculate_mape(outputs, targets):
    outputs = outputs.cpu().detach().numpy().reshape(-1, 1)
    outputs = np.expm1(outputs)
    targets = targets.cpu().detach().numpy().reshape(-1, 1)
    targets = np.expm1(targets)

    epsilon = 1e-8  # Small value to avoid division by zero
    targets_clamped = np.clip(np.abs(targets), a_min=epsilon, a_max=None)
    mape = np.mean(np.abs((targets - outputs) / targets_clamped)) * 100

    return mape

# Training loop
def train_model(model, train_loader, test_loader, num_epochs=20, learning_rate=0.0001, device='cpu',target_name='video_share_count'):
    save_path= f'{target_name}_model.pt'
    criterion = nn.MSELoss()
    optimizer = torch.optim.Adam(model.parameters(), lr=learning_rate)
    model.to(device)

    for epoch in range(num_epochs):
        model.train()
        for feature1, feature2, feature3, feature4, feature5, feature6, targets in train_loader:
            feature1, feature2, feature3, feature4, feature5, feature6 = (
                feature1.to(device), feature2.to(device), feature3.to(device),
                feature4.to(device), feature5.to(device), feature6.to(device)
            )
            #print(feature1.shape, feature2.shape, feature3.shape, feature4.shape, feature5.shape, feature6.shape)
            targets = targets.to(device)

            optimizer.zero_grad()
            outputs = model(feature1, feature2, feature3, feature4, feature5, feature6)
            loss = criterion(outputs, targets)
            loss.backward()
            optimizer.step()

        model.eval()
        test_loss = 0
        total_mape = 0
        num_batches = 0
        with torch.no_grad():
            for feature1, feature2, feature3, feature4, feature5, feature6, targets in test_loader:
                feature1, feature2, feature3, feature4, feature5, feature6 = (
                    feature1.to(device), feature2.to(device), feature3.to(device),
                    feature4.to(device), feature5.to(device), feature6.to(device)
                )
                targets = targets.to(device)
                outputs = model(feature1, feature2, feature3, feature4, feature5, feature6)
                test_loss += criterion(outputs, targets).item()
                mape = calculate_mape(outputs, targets)
                total_mape += mape
                num_batches += 1

        avg_test_loss = test_loss / num_batches if num_batches > 0 else float('inf')
        avg_mape = total_mape / num_batches if num_batches > 0 else float('inf')
        print(f'Epoch: {epoch+1}/{num_epochs}, Test MSE: {avg_test_loss:.4f}, Test MAPE: {avg_mape}')
    torch.save(model.state_dict(), save_path)
    print(f'Model saved at {save_path}')
```

```python
import torch
import csv
import pandas as pd
import numpy as np
import os
from torch.utils.data import Dataset, DataLoader

# Function to load the saved model
def load_saved_model(feature_sizes, device='cpu',target_name='video_share_count'):
    model_path=f'{target_name}_model.pt'
    model = VideoHeartPredictor(feature_sizes)
    model.load_state_dict(torch.load(model_path, map_location=device))  # Load the model state dictionary
    model.to(device)
    model.eval()  # Set the model to evaluation mode
    print(f"Model loaded from {model_path} for prediction.")
    return model


# Prediction function with video_id
def predict_and_save(model, test_loader, device='cpu',target_name='video_share_count'):
    output_csv=f'{target_name}_prediction.csv'
    model.eval()
    predictions = []
    video_ids = []
    with torch.no_grad():
        for feature1, feature2, feature3, feature4, feature5, feature6, _, video_id in test_loader:
            feature1, feature2, feature3, feature4, feature5, feature6 = (
                feature1.to(device), feature2.to(device), feature3.to(device),
                feature4.to(device), feature5.to(device), feature6.to(device)
            )
            #print(feature1.shape, feature2.shape, feature3.shape, feature4.shape, feature5.shape, feature6.shape)
            # Forward pass
            outputs = model(feature1, feature2, feature3, feature4, feature5, feature6)

            # Inverse log transformation and rounding
            outputs = outputs.cpu().numpy().reshape(-1, 1)
            outputs = np.expm1(outputs)  # Inverse of log transformation
            outputs = np.round(outputs).astype(int)

            # Store the rounded predictions and corresponding video IDs
            predictions.extend(outputs)
            video_ids.extend(video_id)

    # Save predictions to CSV
    with open(output_csv, mode='w', newline='') as file:
        writer = csv.writer(file)
        writer.writerow(['video_id', f'{target_name}'])
        for video_id, prediction in zip(video_ids, predictions):
            writer.writerow([video_id, prediction[0]])

    print(f"Predictions saved to '{output_csv}'")
```

```python
class VideoHeartPredictor(nn.Module):
    def __init__(self, feature_sizes,feature=256):
        super(VideoHeartPredictor, self).__init__()

        # Adjusted input sizes based on feature_sizes (e.g., feature_sizes[0] = 1)
        self.fc1_1 = nn.Linear(feature_sizes[0], feature)  # For feature1, which has 1 element
        self.fc1_2 = nn.Linear(feature_sizes[1], feature)  # Adjust for the correct sizes of other features
        self.fc1_3 = nn.Linear(feature_sizes[2], feature)
        self.fc1_4 = nn.Linear(feature_sizes[3], feature)
        self.fc1_5 = nn.Linear(feature_sizes[4], feature)
        self.fc1_6 = nn.Linear(feature_sizes[5], feature)

        # Batch normalization layers for each feature
        self.bn1_1 = nn.BatchNorm1d(feature)
        self.bn1_2 = nn.BatchNorm1d(feature)
        self.bn1_3 = nn.BatchNorm1d(feature)
        self.bn1_4 = nn.BatchNorm1d(feature)
        self.bn1_5 = nn.LayerNorm(feature)
        self.bn1_6 = nn.LayerNorm(feature)

        # Dropout layer after the first layer for regularization
        self.dropout1 = nn.Dropout(0.3)  # 30% dropout rate

        # Combining layers
        self.fc_combined = nn.Linear(feature * 6, 1024)
        self.bn_combined = nn.BatchNorm1d(1024)
        self.dropout2 = nn.Dropout(0.4)  # 40% dropout rate

        # Further dense layers
        self.fc2 = nn.Linear(1024, 128)
        self.bn2 = nn.BatchNorm1d(128)
        self.dropout3 = nn.Dropout(0.2)

        self.fc3 = nn.Linear(128, 1)

        # Apply Kaiming initialization to each layer
        self._initialize_weights()

    def forward(self, feature1, feature2, feature3, feature4, feature5, feature6):
        # Process each feature through its respective layer with batch normalization
        feature1 = F.gelu(self.bn1_1(self.fc1_1(feature1.view(feature1.size(0), -1))))
        feature2 = F.gelu(self.bn1_2(self.fc1_2(feature2.view(feature2.size(0), -1))))
        feature3 = F.gelu(self.bn1_3(self.fc1_3(feature3.view(feature3.size(0), -1))))
        feature4 = F.gelu(self.bn1_4(self.fc1_4(feature4.view(feature4.size(0), -1))))
        feature5 = F.gelu(self.bn1_5(self.fc1_5(feature5.view(feature5.size(0), -1))))
        feature6 = F.gelu(self.bn1_6(self.fc1_6(feature6.view(feature6.size(0), -1))))

        # Concatenate the processed features
        combined = torch.cat([feature1, feature2, feature3, feature4, feature5, feature6], dim=1)

        # Apply dropout and batch normalization after combining features
        combined = F.relu(self.bn_combined(self.fc_combined(combined)))
        combined = self.dropout2(combined)

        # Forward through the second layer with batch normalization and dropout
        combined = F.relu(self.bn2(self.fc2(combined)))
        combined = self.dropout3(combined)

        # Final output
        output = self.fc3(combined)

        return output

    def _initialize_weights(self):
        # Kaiming initialization for the fully connected layers
        for m in self.modules():
            if isinstance(m, nn.Linear):
                nn.init.kaiming_normal_(m.weight, mode='fan_in', nonlinearity='relu')
                if m.bias is not None:
                    nn.init.constant_(m.bias, 0)
```

```python
import pandas as pd
import numpy as np
import re
from collections import Counter
from datetime import datetime
import holidays
import os
from moviepy.editor import VideoFileClip
# Function to extract hashtags and mentions from the video description
def extract_tags(description):
    description = str(description)
    hashtags = re.findall(r'#(\w+)', description)
    mentions = re.findall(r'@(\w+)', description)
    return hashtags, mentions

def total_tag_frequency(tags_list, frequency_dict):
    return sum(frequency_dict[tag] for tag in tags_list if tag in frequency_dict)

# Function to extract date components and check for US holidays
def extract_date_components_and_check_holidays(row):
    date = datetime.utcfromtimestamp(int(row['video_create_date']))
    is_holiday = date in holidays.UnitedStates()
    if 9 <= date.hour < 18:
        time_period = 'Work Time'
    elif 18 <= date.hour < 23:
        time_period = 'Leisure Time'
    else:
        time_period = 'Sleep Time'
    return pd.Series({
        'year': date.year,
        'month': date.month,
        'day': date.day,
        'hour': date.hour,
        'is_holiday': is_holiday,
        'time_period': time_period
    })
# 定义一个函数来获取视频的时长
from moviepy.editor import VideoFileClip
import os
import pandas as pd

# 定义一个函数来获取视频的详细信息
def get_video_details(video_path):
    try:
        clip = VideoFileClip(video_path)
        duration = clip.duration  # 视频时长，单位为秒
        n_frames = clip.reader.nframes  # 视频总帧数
        frame_rate = clip.fps  # 帧率
        width = clip.size[0]  # 视频宽度
        height = clip.size[1]  # 视频高度
        # 检查是否有音频轨道
        clip.close()
        print(duration, n_frames, frame_rate, width, height)
        return duration, n_frames, frame_rate, width, height
    except Exception as e:
        print(e)
        return None, None, None, None, None




# Load the Excel files
training_data = pd.read_excel('/content/drive/MyDrive/training.xlsx', engine='openpyxl')
testing_data = pd.read_excel('/content/drive/MyDrive/testing.xlsx', engine='openpyxl')
validation_data = pd.read_excel('/content/drive/MyDrive/testing_set_with_labels.xlsx', engine='openpyxl')
validation_data = validation_data[['identifier','video_comment_count', 'video_heart_count','video_play_count', 'video_share_count']]
print(validation_data.head())
# 应用函数并将结果保存到新的DataFrame列中
training_data['video_details'] = training_data['video_id'].apply(
    lambda x: get_video_details(os.path.join('/content/training/', f'{x[1:]}.avi'))
)
training_data[['video_duration', 'total_frames', 'frame_rate', 'width', 'height']] = pd.DataFrame(training_data['video_details'].tolist(), index=training_data.index)

# 对测试数据执行相同的操作
testing_data['video_details'] = testing_data['video_id'].apply(
    lambda x: get_video_details(os.path.join('/content/testing/', f'{x[1:]}.avi'))
)
testing_data[['video_duration', 'total_frames', 'frame_rate', 'width', 'height']] = pd.DataFrame(testing_data['video_details'].tolist(), index=testing_data.index)


# 假设 training_data 和 testing_data 是你的DataFrame
# 你需要填补的列列表
columns_to_fill = ['video_duration', 'total_frames', 'frame_rate', 'width', 'height']

# 计算 training_data 中每个 author_id 的中位数
medians = training_data.groupby('author_id')[columns_to_fill].median()

# 使用这些中位数填补 training_data 中的空值
for column in columns_to_fill:
    training_data[column] = training_data.groupby('author_id')[column].transform(lambda x: x.fillna(x.median()))



# 获取video_create_date的最小值和最大值
min_date = min(training_data['video_create_date'].min(), testing_data['video_create_date'].min())
max_date = max(training_data['video_create_date'].max(), testing_data['video_create_date'].max())

# 对training_data和testing_data中的video_create_date进行0-1归一化
training_data['video_create_date_normalized'] = (training_data['video_create_date'] - min_date) / (max_date - min_date)
testing_data['video_create_date_normalized'] = (testing_data['video_create_date'] - min_date) / (max_date - min_date)

# 查看归一化后的数据
print("Training Data after normalization:")
print(training_data[['video_create_date', 'video_create_date_normalized']].head())

print("Testing Data after normalization:")
print(testing_data[['video_create_date', 'video_create_date_normalized']].head())
# training_data['video_duration'] = training_data['video_id'].apply(
#     lambda x: get_video_duration(os.path.join('/content/training', f'{x[1:]}.avi'))
# )
# testing_data['video_duration'] = testing_data['video_id'].apply(
#     lambda x: get_video_duration(os.path.join('/content/testing', f'{x[1:]}.avi'))
# )


# Extract hashtags and mentions and assign them to new columns
training_data['hashtags'], training_data['mentions'] = zip(*training_data['video_description'].apply(extract_tags))
testing_data['hashtags'], testing_data['mentions'] = zip(*testing_data['video_description'].apply(extract_tags))

# Combine and flatten tags from both datasets
all_hashtags = [tag for sublist in training_data['hashtags'].tolist() + testing_data['hashtags'].tolist() for tag in sublist]
all_mentions = [tag for sublist in training_data['mentions'].tolist() + testing_data['mentions'].tolist() for tag in sublist]

# Count the frequency of each hashtag and mention
hashtag_frequency = Counter(all_hashtags)
mention_frequency = Counter(all_mentions)

# Calculate total frequency for each sample and add as new columns
training_data['total_hashtag_frequency'] = training_data['hashtags'].apply(lambda tags: total_tag_frequency(tags, hashtag_frequency))
testing_data['total_hashtag_frequency'] = testing_data['hashtags'].apply(lambda tags: total_tag_frequency(tags, hashtag_frequency))

training_data['total_mention_frequency'] = training_data['mentions'].apply(lambda tags: total_tag_frequency(tags, mention_frequency))
testing_data['total_mention_frequency'] = testing_data['mentions'].apply(lambda tags: total_tag_frequency(tags, mention_frequency))
training_data['hashtag_count'] = training_data['hashtags'].apply(len)
testing_data['hashtag_count'] = testing_data['hashtags'].apply(len)

training_data['mention_count'] = training_data['mentions'].apply(len)
testing_data['mention_count'] = testing_data['mentions'].apply(len)
# Apply functions to extract date components and check holidays
training_data[['year', 'month', 'day', 'hour', 'is_holiday', 'time_period']] = training_data.apply(extract_date_components_and_check_holidays, axis=1)
testing_data[['year', 'month', 'day', 'hour', 'is_holiday', 'time_period']] = testing_data.apply(extract_date_components_and_check_holidays, axis=1)

# Define the columns on which to apply the log transformation
columns_to_log = [
    'author_follower_count', 'author_following_count',
    'author_total_heart_count', 'author_total_video_count',
    #'video_comment_count', 'video_heart_count',
    #'video_play_count', 'video_share_count'
]

# Apply log1p to the specified columns in both training and testing data
for column in columns_to_log:
    training_data[column] = np.log1p(training_data[column])
    testing_data[column] = np.log1p(testing_data[column])
    #validation_data[column] = np.log1p(validation_data[column])

columns_to_drop = [
    'video_id', 'video_description', 'video_definition',
    'video_format', 'hashtags', 'mentions',
]

# 从每个数据集中删除这些列
testing_data.drop(columns=columns_to_drop, errors='ignore', inplace=True)
training_data.drop(columns=columns_to_drop, errors='ignore', inplace=True)

# Display some of the updated data to verify
print("Transformed Training Data Sample:")
print(training_data.head())
print("\nTransformed Testing Data Sample:")
print(testing_data.head())
validation_data = pd.merge(testing_data, validation_data, on='identifier', how='inner')
print("\nTransformed Validation Data Sample:")
print(validation_data.head())
validation_data.to_csv('validation_data.csv', index=False)
training_data.to_csv('training_data.csv', index=False)
testing_data.to_csv('testing_data.csv', index=False)
```

```python
import pandas as pd
import numpy as np
from catboost import CatBoostRegressor
from xgboost import XGBRegressor
from lightgbm import LGBMRegressor
from lce import LCERegressor
from sklearn.metrics import mean_squared_error, mean_absolute_percentage_error
# 要绘制的列
columns_to_plot = ['video_comment_count', 'video_heart_count', 'video_play_count', 'video_share_count']

# 定义删除异常值的函数，使用IQR（四分位距）
def remove_outliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]
# 加载数据
training_data = pd.read_csv('/content/training_data.csv')
testing_data = pd.read_csv('/content/testing_data.csv')
validation_data = pd.read_csv('/content/validation_data.csv')
for col in columns_to_plot:
  training_data= remove_outliers(training_data, col)
# 删除 'identifier' 列（如果有必要）
for df in [training_data, testing_data, validation_data]:
    if 'identifier' in df.columns:
        df.drop(columns=['identifier'], inplace=True)

# 转换分类变量
categorical_features = ['author_id', 'time_period', 'is_holiday', 'year', 'month',]
#categorical_features = ['author_id', 'time_period', 'is_holiday']
training_data[categorical_features] = training_data[categorical_features].astype('category')
validation_data[categorical_features] = validation_data[categorical_features].astype('category')
testing_data[categorical_features] = testing_data[categorical_features].astype('category')

# 定义特征和目标
features = [
    'author_id', 'author_follower_count', 'author_following_count','hashtag_count','mention_count',
    'author_total_heart_count', 'author_total_video_count',"video_create_date_normalized",
    'total_hashtag_frequency', 'total_mention_frequency', 'year', 'month','day', 'hour',
    'is_holiday', 'time_period','video_duration', 'total_frames', 'frame_rate', 'width', 'height'
]
targets = [
    'video_comment_count', 'video_heart_count', 'video_play_count', 'video_share_count'
]

# 对训练和验证数据的目标变量应用对数变换
for target in targets:
    training_data[target] = np.log1p(training_data[target])
    validation_data[target] = np.log1p(validation_data[target])

# 定义训练集和验证集
X_train = training_data[features]
y_train = training_data[targets]
X_valid = validation_data[features]
y_valid = validation_data[targets]
# 使用pandas的get_dummies函数对分类特征进行独热编码
X_train = pd.get_dummies(X_train, columns=categorical_features)
X_valid = pd.get_dummies(X_valid, columns=categorical_features)
X_test = pd.get_dummies(testing_data[features], columns=categorical_features)
from sklearn.preprocessing import MinMaxScaler

# 初始化 MinMaxScaler
scaler = MinMaxScaler()

# 对训练集进行拟合并转换
X_train = scaler.fit_transform(X_train)

# 使用相同的缩放器对验证集和测试集进行转换（不再拟合）
X_valid = scaler.transform(X_valid)
X_test = scaler.transform(X_test)

# 计算分类特征索引
#categorical_features_indices = [X_train.columns.get_loc(c) for c in categorical_features]

# 初始化模型
from pytorch_tabnet.tab_model import TabNetRegressor
from sklearn.linear_model import (
    LinearRegression, Ridge, Lasso, ElasticNet, BayesianRidge, HuberRegressor,
    SGDRegressor, PassiveAggressiveRegressor, RANSACRegressor, TheilSenRegressor, ARDRegression
)
from sklearn.ensemble import (
    RandomForestRegressor, GradientBoostingRegressor, AdaBoostRegressor,
    ExtraTreesRegressor, BaggingRegressor, VotingRegressor, StackingRegressor
)
from sklearn.svm import SVR, NuSVR, LinearSVR
from sklearn.neighbors import KNeighborsRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.gaussian_process import GaussianProcessRegressor
from sklearn.neural_network import MLPRegressor
from sklearn.isotonic import IsotonicRegression
from sklearn.cross_decomposition import PLSRegression
models = {
    'CatBoost': CatBoostRegressor(verbose=0, task_type="CPU"),
    'XGBoost': XGBRegressor(enable_categorical=True, task_type="CPU", tree_method="hist"),
    'LightGBM': LGBMRegressor(),
    'lce_model': LCERegressor(random_state=817),
    # Sklearn models
    'LinearRegression': LinearRegression(),
    'Ridge': Ridge(),
    'Lasso': Lasso(),
    'ElasticNet': ElasticNet(),
    'BayesianRidge': BayesianRidge(),
    'HuberRegressor': HuberRegressor(),
    'ARDRegression': ARDRegression(),
    'SGDRegressor': SGDRegressor(),
    'PassiveAggressiveRegressor': PassiveAggressiveRegressor(),
    #'RANSACRegressor': RANSACRegressor(),
    #'TheilSenRegressor': TheilSenRegressor(),

    'RandomForestRegressor': RandomForestRegressor(random_state=817),
    'GradientBoostingRegressor': GradientBoostingRegressor(random_state=817),
    'AdaBoostRegressor': AdaBoostRegressor(random_state=817),
    'ExtraTreesRegressor': ExtraTreesRegressor(random_state=817),
    'BaggingRegressor': BaggingRegressor(random_state=817),
    'SVR': SVR(),
    'NuSVR': NuSVR(),
    'LinearSVR': LinearSVR(),

    'KNeighborsRegressor': KNeighborsRegressor(),
    'DecisionTreeRegressor': DecisionTreeRegressor(random_state=817),

    'GaussianProcessRegressor': GaussianProcessRegressor(),
    'MLPRegressor': MLPRegressor(random_state=817),

    'PLSRegression': PLSRegression(),

    # TabNet
    'TabNet': TabNetRegressor(verbose=0),
}


# 创建一个空的 DataFrame 来保存测试集的预测结果
test_predictions = pd.DataFrame()
results = pd.DataFrame()

# 训练模型并进行预测
for target in targets:
    y_train_target = y_train[target]
    y_valid_target = y_valid[target]  # 验证集目标

    for name, model in models.items():
        print(f"Training {name} on {target}...")
        # 训练模型
        if name == 'TabNet':
          y_train_target = y_train_target.values.reshape(-1, 1)
          y_valid_target = y_valid_target.values.reshape(-1, 1)
        model.fit(X_train, y_train_target)

        # 在验证集上预测
        y_pred_log_valid = model.predict(X_valid)
        y_pred_valid = np.expm1(y_pred_log_valid)  # 逆对数变换还原

        # 计算 MAPE 和 MSE
        mse = mean_squared_error(np.expm1(y_valid_target), y_pred_valid)
        mape = mean_absolute_percentage_error(np.expm1(y_valid_target), y_pred_valid)

        # 将结果保存到 DataFrame
        results = pd.concat([results, pd.DataFrame({'Model': [name], 'Target': [target], 'MSE': [mse], 'MAPE': [mape]})])

        # 在测试集上预测
        y_pred = np.expm1(model.predict(X_valid))
        test_predictions[f'{name}_{target}'] = y_pred

# 输出验证集上的性能指标
print(results)
results.to_csv('/content/results.csv', index=False)
# 保存测试集的预测结果
test_predictions.to_csv('/content/test_predictions.csv', index=False)
print("Predictions saved to CSV.")
#X_train.to_csv('/content/X_train.csv', index=False)
#X_valid.to_csv('/content/X_valid.csv', index=False)
#X_test = testing_data[features]
#X_test.to_csv('/content/X_test.csv', index=False)
```
